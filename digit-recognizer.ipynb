```python
train_df = pd.read_csv('/kaggle/input/digit-recognizer/train.csv')
train_df2 = train_df.drop(columns = 'label')
train_arr = np.array(train_df2)
X = train_arr.T
y_data = np.array(train_df)
y_data=y_data.T
Y = y_data[0]
X
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.losses import SparseCategoricalCrossentropy
```

```python
model = Sequential([Dense(128, activation='relu', input_shape=(784,)), Dense(units = 10, activation = 'sigmoid'), Dense(units = 10, activation = 'softmax')])
```

```python
model.compile(loss = 'sparse_categorical_crossentropy')
model.fit(X.T, Y, epochs = 100)
```

```python
df=pd.read_csv('/kaggle/input/digit-recognizer/test.csv')
x = np.array(df)
z=model.predict(x)
```

```python
array=[]
for i in range(len(z)):
    maximum = 0
    index=0
    for j in range(10):
        if maximum<z[i][j]:
            maximum = z[i][j]
            index=j
    array.append(index)
array
```
